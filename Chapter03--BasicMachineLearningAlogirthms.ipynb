```python
%load_ext autoreload
%autoreload
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from missingmap import missmap
```

```python
df = pd.read_excel("dds_datasets/rollingsales_manhattan.xls", 0, skiprows=4)
df = df.rename(columns=lambda x: x.replace('\n', '').replace('-', '_').replace(' ', '_').lower())
df['neighborhood'] = df.neighborhood.str.strip()
df['building_class_category'] = df.building_class_category.str.strip()

neighborhood, neighborhood_keys = pd.factorize(df.neighborhood)
df['neighborhood_f'] = neighborhood

building_f, building_keys = pd.factorize(df.building_class_category)
df['building_f'] = building_f

df[['saleprice_ln',
    'gross_square_feet_ln',
    'land_square_feet_ln']] = np.log(df[['saleprice', 'gross_square_feet',
                                         'land_square_feet']])
```

```python
pure = df.query("saleprice > 0 & gross_square_feet > 0 & land_square_feet > 0")
```

```python
pure[['saleprice', 'gross_square_feet', 'land_square_feet']].describe()
```

```python
mod = sm.OLS.from_formula("saleprice_ln ~ gross_square_feet_ln", pure)
res = mod.fit()
res.summary()
```

```python
# not random noise in the residuals
# what is that large cluster around 11.75 gsqrf_ln and 12 saleprice?

sns.regplot('gross_square_feet_ln', 'saleprice_ln', pure)
```

```python
fig, ax = plt.subplots()
ax.scatter(res.fittedvalues, res.resid, color='k', alpha=.7)
```

```python
mod2 = sm.OLS.from_formula("saleprice_ln ~ land_square_feet_ln + gross_square_feet_ln +"
                           "neighborhood_f", pure)
res2 = mod2.fit()
res2.summary()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 8))
ax1.scatter(res.fittedvalues, res.resid, color='k', alpha=.7)
ax1.set_title("Model 1")
ax2.scatter(res2.fittedvalues, res2.resid, color='k', alpha=.7)
ax2.set_title("Model 2")
ax2.set_xlim(ax1.get_xlim())
```

```python
mod3 = sm.OLS.from_formula("saleprice_ln ~ land_square_feet_ln + gross_square_feet_ln +"
                           "neighborhood_f + building_f", pure)
res3 = mod3.fit()
res3.summary()
```

```python
fig, ax = plt.subplots()
ax.scatter(res3.fittedvalues, res3.resid, color='k', alpha=.7)
```

```python
ax = sns.distplot(res2.resid)
ax.set_title("Model 2 Residuals")
```

```python
# interact neighborhood and building category
mod4 = sm.OLS.from_formula("saleprice_ln ~ gross_square_feet_ln + land_square_feet_ln + "
                           "neighborhood_f * building_f", pure)
res4 = mod4.fit()
res4.summary()
```

```python
ax = sns.distplot(res4.resid)
ax.set_title("Model 4 Residuals")

sns.regplot(res4.fittedvalues, res4.resid)
```

```python
sns.distplot(pure.saleprice_ln)
```

```python
# let's punt a bit on the reason, and just make a dummy for < 13
pure['saleprice_lt_d'] = pure.saleprice_ln < 13
```

```python
mod5 = sm.OLS.from_formula("saleprice_ln ~ land_square_feet_ln + gross_square_feet_ln +"
                           "neighborhood_f + building_f + saleprice_lt_d", pure)
res5 = mod5.fit()
res5.summary()
```

```python
sns.distplot(res5.resid)
sns.regplot(res5.fittedvalues, res5.resid)
```

That boosted the $R^2$ a lot. It really does seem like there are two samples here.

## With K-NN from scikit-learn

```python
from sklearn.neighbors import NearestNeighbors
```

```python
y = pure.saleprice_ln.values
features = ['neighborhood_f', 'gross_square_feet_ln',
           'land_square_feet_ln', 'residential_units', 'commercial_units',
           'total_units', 'year_built']
X = pure[features].values
```

```python
nbrs = NearestNeighbors(n_neighbors=2, algorithm='ball_tree').fit(X)
distances, indices = nbrs.kneighbors(X) 
```

## Ridge Regression

```python
from sklearn import linear_model
clf = linear_model.Ridge(alpha=.5)
clf.fit(X, y)
zip(features, clf.coef_)
```

```python
n_alphas = 200
alphas = np.logspace(-10, -2, n_alphas)
clf = linear_model.Ridge(fit_intercept=True)

coefs = []
for a in alphas:
    clf.set_params(alpha=a)
    clf.fit(X, y)
    coefs.append(clf.coef_)

fig, ax = plt.subplots()
ax.set_color_cycle(['b', 'r', 'g', 'c', 'k', 'y', 'm'])

ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
ax.set_xlabel('alpha')
ax.set_ylabel('weights')
ax.set_title('Ridge coefficients as a function of the regularization')
```

## Lasso

```python
clf = linear_model.Lasso(alpha = 0.1)
clf.fit(X, y)
zip(features, clf.coef_)
```
