We can now [`groupby`](http://pandas.pydata.org/pandas-docs/stable/groupby.html) the `age_bin` and give some summary statistics.

A lot of those ages don't look valid, we have 137106 users at age 0 or lower. Let's see what's going on.

````python
pd.value_counts(df.Age).head()
````

So it looks like a bunch of users have their age set to zero. Presumably the same thing is happening with gender too.

People could be lying about their age, but that's a lot of people. It could be that 0 is the default value given if you're not signed in to your NYT account. If you didn't sign up, you never provided your age (or gender). Let's test that hypothesis by grouping by whether or not you're signed in. If everyone who is not signed in has an age of 0, then we'll assume we're correct.

````python
df.groupby("Signed_In").mean()
````

Yep, that looks like it. So if someone isn't signed in they're given an Age of 0 and a Gender of female. Those *should* be `NaNs` (Not a Number), but as it is we'll be careful about our analysis and filter to only signed in users when reporting Age or Gender statsitics.

````python
df.groupby("age_bin").mean()
````

Next we want to "Define a new variable to segment or categorize users based on their click behavior." This is a bit vague, so let's take a look at the distribution of clicks before doing any binning.

````python
pd.value_counts(df.Clicks).sort_index()
````

So most people are at 0 clicks. And about an order of magnitude between the counts for each click.

Now we want to calculate the Click Through Rate (CTR) = $ \frac{|Clicks|}{|Impressions|} $.

We'll need to filter out the users with zero impressions so we aren't dividing by 0 anywhere. `NaN`s are created in those locations for `ctr`.

````python
idx = df[df.Impressions > 0].index  # avoind dividing by infinity below
df['ctr'] = df.loc[idx, 'Clicks'] / df.loc[idx, 'Impressions']
````

````python
df.groupby(df.Impressions > 0)['Clicks'].describe().unstack()
````

````python
# Some Density plots:

axes = df[df.Impressions > 0].groupby('age_bin')['ctr'].plot(kind='kde')
````

That density plot isn't all that helpful since so much of the mass is concentrated at 0. Let's filter those out.

````python
# again we have the label bug. Someone should fix that.

fig, ax = plt.subplots()
groups = df[df.Clicks > 0].groupby('age_bin')['ctr']

for k, df_g in groups:
    df_g.plot(kind='kde', ax=ax, label=k)

ax.set_xlabel("Click Through Rate")
ax.legend()
````

And we can define a function to plot the distribution of impressions and the CTR, without grouping by age.

````python
# Distributions impression and CTR
def plot_distributions(df):
    
    fig, (ax1, ax2) = plt.subplots(ncols=2)
    
    ax1 = df.Impressions.hist(bins=20, color='k', ax=ax1)
    ax1.set_title("Impressions")
    
    ax2 = df.ctr.hist(bins=20, ax=ax2, log=True)
    ax2.set_title("Log Click-Through-Rate")
    return ax1, ax2
````

````python
plot_distributions(df)
````

So that's it for a single day. At this point, we want to generalize what we've done to other days.

This is a pretty common pattern in EDA. Take a (hopefully representative) subsample of your data, explore it all the way through, and then refactor your code to handle new samples. Of course, you'll think of new things to do along the way, so it's not ever that straightforward, it's more of a cycle between exploration and generalization.

The API we'll implement will take a filename as an input. The output will be a set of graphs and tables. For now we'll just display them in the notebook, but you'd probably want to save them to disk.

````python
def explore(infile):
    """
    Perform some EDA on a nyt add file.
    
    Parameters
    ----------    
    infile: str
        path to a csv with (simulated) NYT add tracking
        
    Returns
    -------
    None

    """
    df = pd.read_csv(infile)
    bins = [-np.inf, 0, 17, 24, 34, 44, 54, 64, np.inf]
    df['age_bin'] = pd.cut(df.Age, bins)
    # for identifying plots
    df.name = infile.split('/')[-1].split('.')[0]
    
    print(describe_age_by_age_bin(df))
    print(means_by_signed_in(df))
    print(means_by_bin(df))
    
    plot_impressions_by_age(df)
    plot_impressions_boxplot(df)
    
    # CTR
    idx = df[df.Impressions > 0].index
    df['ctr'] = df.loc[idx, 'Clicks'] / df.loc[idx, 'Impressions']
    
    print(describe_clicks_by_impressions(df))
    plot_ctr_by_age(df)
    plot_distributions(df)
    
def describe_age_by_age_bin(df):
    return df.groupby(df.age_bin)['Age'].agg([len, np.min, np.max, np.mean])

def means_by_signed_in(df):
    return df.groupby("Signed_In").mean()

def means_by_bin(df):
    return df.groupby("age_bin").mean()

def plot_impressions_by_age(df):
    groups = df.groupby("age_bin")['Impressions']
    fig, ax = plt.subplots()

    for k, v in groups:
        v.hist(label=k, alpha=.75, ax=ax)

    ax.legend()
    ax.set_xlabel("Impressions")
    ax.set_title(df.name)
    plt.draw()
    plt.show()
    return ax

def plot_impressions_boxplot(df):
    ax = df.boxplot("Impressions", by='age_bin')
    ax.set_title("Impressions for " + df.name)
    return ax

def describe_clicks_by_impressions(df):
    return df.groupby(df.Impressions > 0)['Clicks'].describe().unstack()

def plot_ctr_by_age(df):
    # again we have the label bug. Someone should fix that.
    fig, ax = plt.subplots()
    groups = df[df.Clicks > 0].groupby('age_bin')['ctr']

    for k, df_g in groups:
        df_g.plot(kind='kde', ax=ax, label=k)

    ax.legend()
    ax.set_xlabel("Click Through Rate")
    ax.set_title(df.name)
    plt.draw()
    plt.show()
    return ax

def plot_distributions(df):
    
    fig, (ax1, ax2) = plt.subplots(ncols=2)
    
    ax1 = df.Impressions.hist(bins=20, color='k', ax=ax1)
    ax1.set_title("Impressions for " + df.name)
    
    ax2 = df.ctr.hist(bins=20, ax=ax2, log=True)
    ax2.set_title("Log Click-Through-Rate for " + df.name)
    return ax1, ax2
````

All we've done is taken the code we did manually earlier, and put them into functions.
Every function takes a `DataFrame` as it's input and returns either a `DataFrame` or a matplotlib `axes`.
Any function that makes a plot is prefixed by `plot_` and then a description of what it plots (good, *consistent* naming is important).

We should be able to call `explore` on 'dds_datasets/nyt1.csv' and get back pretty much what we've done up until this point. We need to be able to tell which dataset is attached to which plot, so I've attached those to `df.name` and the plot titles.

````python
explore('dds_datasets/nyt1.csv')
````

Pretty cool. We should probably do some more labeling, but that works for now. Let's iterate over a few files:

````python
files = ['dds_datasets/nyt2.csv', 'dds_datasets/nyt3.csv', 'dds_datasets/nyt4.csv', 'dds_datasets/nyt5.csv']

for f in files:
    explore(f)
````
