```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
df = pd.read_csv("dds_datasets/rollingsales_manhattan_cleaned.csv")
df[['saleprice', 'gross_square_feet']] = df[['saleprice', 'gross_square_feet']].replace(0, np.nan)
df.head()
```

```python
neighborhood, neighborhood_keys = pd.factorize(df.neighborhood)
df['neighborhood_f'] = neighborhood
```

```python
mod = sm.OLS.from_formula("np.log(saleprice) ~ np.log(gross_square_feet)", df.dropna())
res = mod.fit()
res.summary()
```

```python
sub = np.log(df[['saleprice', 'gross_square_feet', 'land_square_feet']]).dropna()
sub_idx = sub.index
sns.regplot('gross_square_feet', 'saleprice', sub)
```

```python
sub2 = pd.concat([sub, df[['neighborhood_f']]], axis=1).ix[sub.index]
sub2 = sub2.replace([-np.inf, np.inf], [np.nan, np.nan]).dropna()
mod2 = sm.OLS.from_formula("saleprice ~ land_square_feet + gross_square_feet + neighborhood_f", sub2)
res2 = mod2.fit()
res2.summary()
```

```python
fig, ax = plt.subplots()
sm.graphics.plot_fit(res2, 'land_square_feet', ax=ax)
```

```python
sns.distplot(res2.resid, axlabel="Model 2 Residuals")
```
