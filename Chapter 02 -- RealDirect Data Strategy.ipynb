```python
# The first 4 rows are some header info. Just discard for now.

df = pd.read_excel("dds_datasets/rollingsales_manhattan.xls", 0,
                   skiprows=4)
print(df.dtypes)
df.head()
```

Building classifications: http://nycserv.nyc.gov/nycproperty/help/hlpbldgcode.html

```python
# more pythonic
df = df.rename(columns=lambda x: x.replace('\n', '').replace('-', '_').replace(' ', '_').lower())
```

```python
# massive outliers in sale price
# TODO: matplotlib offset xticklabels

ax = df.saleprice.hist(log=True, bins=50, color='k', label='Sale Price')
ax.set_ylabel('Log Count')
ax.legend()
```

```python
# There's a lot of sales at 0:

pd.value_counts(df.saleprice).head()
```

```python
df_sold = df.query('saleprice > 0')  # query added in .13
ax = df_sold.saleprice.hist(bins=50, log=True)
ax.set_xlabel("Square Feet")
```

```python
# Plotting the counts of the log price may be better
# than the log count of the price. The interpretation is
# a bit trickier though :/

ax = pd.Series(np.log(df_sold.saleprice)).hist(color='k', bins=40)
ax.set_xlabel("Log Sale Price")
```

```python
ax = df.gross_square_feet.hist(bins=50, color='k', log=True)
ax.set_title("Log Histogram of Square Footage")
```

```python
# Simple scatter of price, sqr foot relationship
```

```python
ax = df_sold.plot(kind='scatter', x='gross_square_feet', y='saleprice')
ax.set_xscale('log')
ax.set_xlim(1000.0, 1000000.0)
```

```python
# have fun interpreting.

ax = df_sold.plot(kind='scatter', x='gross_square_feet', y='saleprice')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1000.0, 1000000.0)
# ax.get_xlim()
```

```python
df['year_built'] = df.year_built.replace(0, np.nan)
df.to_csv('dds_datasets/rollingsales_manhattan_cleaned.csv')
```

```python
# same for # of residental units
ax = df.residential_units.hist(bins=30, log=True, alpha=.6, label='residental')
ax = df.commercial_units.hist(bins=30, log=True, alpha=.6, label='commercial', ax=ax)
ax.legend()
```

```python
ax = df.land_square_feet.hist(bins=20, log=True, alpha=.75, color='k', label='land square feet')
df.gross_square_feet.hist(bins=40, log=True, alpha=.5, ax=ax, label='gross square feet')
ax.legend()
```

```python
df[df.year_built > 1000].plot(kind='scatter', x='year_built', y='saleprice')
```

```python
ax = df.groupby('sale_date')['saleprice'].mean().resample('M').plot()
ax.set_ylabel("Average Sale Price")
```
