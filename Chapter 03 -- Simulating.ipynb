```python
import numpy as np
import seaborn as sns
```

```python
x_1 = np.random.normal(loc=5, scale=7, size=1000)

true_error = np.random.normal(0, 2, 1000)
true_beta0 = 1.1
true_beta1 = -8.2
```

```python
sns.distplot(x_1)
```

```python
def mk_y(b0, b1, x, e):
    return b0 + b1 * x + e

y = mk_y(true_beta0, true_beta1, x_1, true_error)
sns.distplot(y, color=sns.desaturate("indianred", .75))
```

```python
# need to transpose, so index are the cols
df = pd.DataFrame([x_1, y], index=['x1', 'y']).T
df.head()
```

```python
model = sm.OLS.from_formula('y ~ x1', df)
res = model.fit()
res.summary()
```

Now $x_2 \widetilde{} \Gamma(k, \theta)$, I'll pick $k=4$, $\theta = 2$. The angels have come down and whispered in our ear that the True underlying CDF is $y = x_1 + x_2$, i.e. still linear. Fit $y$ on both $x$'s separately and then together.

```python
from scipy import stats

df['x2'] = stats.gamma(2, scale=2).rvs(1000)
```

```python
# Model 1 is the same as above
mod1, res1 = model, res
mod2 = sm.OLS.from_formula('y ~ x2', df)
res2 = mod2.fit()

mod_both = sm.OLS.from_formula('y ~ x1 + x2', df)
res_both = mod_both.fit()
```

```python
for res in [res1, res2, res_both]:
    print("{}\n\n".format(res.params))
```

```python
import matplotlib.pyplot as plt
sns.regplot('x1', 'y', df)
sns.regplot('x2', 'y', df)
```

```python
mses = []
sizes = [100, 1000, 10000, 100000]
for size_ in sizes:
    x_1 = np.random.normal(loc=5, scale=7, size=size_)
    true_error = np.random.normal(0, 2, size_)
    df = pd.DataFrame({'x_1': x_1,
                       'e': true_error,
                       'x_2': stats.gamma(2, scale=2).rvs(size_),
                       'y': mk_y(true_beta0, true_beta1, x_1, true_error)})
    mod = sm.OLS.from_formula('y ~ x_1 + x_2', df)
    res = mod.fit()
    mses.append(res.mse_resid)

fig, ax = plt.subplots()
ax.bar(np.log(sizes), mses)  # Should do deviate from mean.
```

```python
mses = []
sizes = [100, 1000, 10000, 100000]
for size_ in sizes:
    x_1 = np.random.normal(loc=5, scale=7, size=size_)
    true_error = np.random.normal(0, 2, size_)
    df = pd.DataFrame({'x_1': x_1,
                       'e': true_error,
                       'x_2': stats.gamma(2, scale=2).rvs(size_),
                       'z': x_1 ** 2,
                       'y': mk_y(true_beta0, true_beta1, x_1, true_error)})
    mod = sm.OLS.from_formula('y ~ x_1 + x_2 + z', df)
    res = mod.fit()
    mses.append(res.mse_resid)

fig, ax = plt.subplots()
ax.bar(np.log(sizes), mses)  # Should do deviate from mean.
```
