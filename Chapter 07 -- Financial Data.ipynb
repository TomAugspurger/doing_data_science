```python
%matplotlib inline

import pandas as pd
from pandas.io.data import DataReader
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = DataReader("AAPL", data_source="yahoo", start="2004-01-01", end="2014-01-01")
```

```python
df.head()
```

```python
# daily log returns

returns_ln = df.Close.div(df.Close.shift())
ax = returns_ln.plot(label='Log Returns')
_ = ax.legend()
```

```python
log_volume = df.Volume.div(df.Volume.shift())
log_volume.plot()
```

```python
# Linear regression with lag of 2 weeks.
import statsmodels.tsa.api as smt

mod = smt.AR(df.Close)
res = mod.fit(14)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 8))
ax1.scatter(df.Close.iloc[14:], res.fittedvalues)

ax2.scatter(res.resid, res.fittedvalues)
```
