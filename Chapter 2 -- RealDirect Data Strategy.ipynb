```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("dds_datasets/rollingsales_manhattan.xls", 0, skiprows=4)  # just 1 sheet
print(df.dtypes)
df.head()
```

Building classifications: http://nycserv.nyc.gov/nycproperty/help/hlpbldgcode.html

```python
df = df.rename(columns=lambda x: x.replace('\n', '').replace('-', '_').replace(' ', '_').lower())
```

```python
df['year_built'] = df.year_built.replace(0, np.nan)
df.to_csv('dds_datasets/rollingsales_manhattan_cleaned.csv')
```

```python
# massive outliers in sale price
ax = df.saleprice.hist(log=True, color='k', label='Sale Price')
ax.legend()
```

```python
# same for # of residental units
ax = df.residential_units.hist(bins=30, log=True, alpha=.6, label='residental')
ax = df.commercial_units.hist(bins=30, log=True, alpha=.6, label='commercial', ax=ax)
ax.legend()
```

```python
ax = df.land_square_feet.hist(bins=20, log=True, alpha=.75, color='k', label='land square feet')
df.gross_square_feet.hist(bins=40, log=True, alpha=.5, ax=ax, label='gross square feet')
ax.legend()
```

```python
df[df.year_built > 1000].plot(kind='scatter', x='year_built', y='saleprice')
```

```python
ax = df.groupby('sale_date')['saleprice'].mean().resample('M').plot()
ax.set_ylabel("Average Sale Price")
```

```python
pd.value_counts(df.year_built).sort_index()
```

```python
df.neighborhood.unique()
```

```python
features = ['residential_units', 'commercial_units', 'total_units', 'land_square_feet',
            'gross_square_feet', 'year_built', 'saleprice']
df.groupby('neighborhood')[features].mean()
```

```python
df.groupby('residential_units')['saleprice'].mean().loc[1:4].plot(kind='barh')
```

```python
df.query('1 <= residential_units <= 4').plot(kind='scatter', x='land_square_feet', y='saleprice')
```

```python
q = '1 <= residential_units <= 4 & 0 < land_square_feet < 1500 & saleprice < 120000000'
df.query(q).plot(kind='scatter', x='land_square_feet', y='saleprice')
```
